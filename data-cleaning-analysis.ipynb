```python
import random 

 

contents = ["Film", "Dizi", "Belgesel", "Muzik"] 

devices = ["Mobile", "TV", "Web", "Tablet"] 

 

with open("new_logs.csv", "w", encoding="utf-8") as f: 

    for i in range(50): 

        user_id = random.randint(1000, 9999) 

        content = random.choice(contents + ["film"]) 

        watch_time = random.choice([10, 25, 45, 80, "", "abc", 100]) 

        device = random.choice(devices) 

 

        row = f"{user_id},{content},{watch_time},{device}" 

        f.write(row + "\n") 

 

print("Yeni veri dosyası oluşturuldu.")
```

```python
import os #klasörü oluşturma ve dizin için

 

os.makedirs("lakehouse/quarantine", exist_ok=True) 

os.makedirs("lakehouse/silver", exist_ok=True) 

 

clean_rows = [] 

bad_rows = [] 

 

valid_contents = ["Film", "Dizi", "Belgesel", "Muzik"] 

 

with open("lakehouse/bronze/new_logs.csv", "r", encoding="utf-8") as f: 

    for line in f: 

        parts = line.strip().split(",") 

 

        content = parts[1] 

        watch_time = parts[2] 

 

        if content not in valid_contents or watch_time == "" or not watch_time.isdigit(): 

            bad_rows.append(line.strip()) 

        else: 

            clean_rows.append(line.strip()) 

 

with open("lakehouse/quarantine/bad_logs.csv", "w", encoding="utf-8") as f: 

    for row in bad_rows: 

        f.write(row + "\n") 

 

with open("lakehouse/silver/clean_new_logs.csv", "w", encoding="utf-8") as f: 

    for row in clean_rows: 

        f.write(row + "\n") 

 

print("Temiz ve bozuk veriler ayrıldı.") 
print("Bozuk kayıt sayısı:",len(bad_rows))
print("Temiz kayıt sayısı:",len(clean_rows))
```

```python
all_clean_rows = [] 

 

# eski silver verisi 

with open("lakehouse/silver/clean_logs.csv", "r", encoding="utf-8") as f: 

    for line in f: 

        all_clean_rows.append(line.strip()) 

 

# yeni temiz silver verisi 

with open("lakehouse/silver/clean_new_logs.csv", "r", encoding="utf-8") as f: 

    for line in f: 

        all_clean_rows.append(line.strip()) 

 

with open("lakehouse/silver/all_clean_logs.csv", "w", encoding="utf-8") as f: 

    for row in all_clean_rows: 

        f.write(row + "\n") 

 

print("Silver katmanı güncellendi.")
```

```python
device_counts = {} 

content_counts = {} 

 

with open("lakehouse/silver/all_clean_logs.csv", "r", encoding="utf-8") as f: 

    for line in f: 

        parts = line.strip().split(",") 

 

        content = parts[1] 

        device = parts[3] 

 

        content_counts[content] = content_counts.get(content, 0) + 1 

        device_counts[device] = device_counts.get(device, 0) + 1 

 

with open("lakehouse/gold/dashboard_report.csv", "w", encoding="utf-8") as f: 

    f.write("type,name,count\n") 

 

    for k, v in content_counts.items(): 

        f.write(f"content,{k},{v}\n") 

 

    for k, v in device_counts.items(): 

        f.write(f"device,{k},{v}\n") 

 

print("Dashboard verisi hazırlandı.") 
```
