```python
import random 

 

contents = ["Film", "Dizi", "Belgesel", "Muzik"] 

devices = ["Mobile", "TV", "Web"] 

 

with open("logs.csv", "w", encoding="utf-8") as f: 

    for i in range(200): 

        user_id = random.randint(1000, 9999) 

        content = random.choice(contents) 

        watch_time = random.randint(5, 120) 

        device = random.choice(devices) 

        row = f"{user_id},{content},{watch_time},{device}" 

        f.write(row + "\n") 

 

print("Ham log verisi oluşturuldu.") 
```

```python
import os 

import shutil 

 

os.makedirs("data_lake/raw", exist_ok=True) 

 

shutil.copy("logs.csv", "data_lake/raw/logs.csv") 

 

print("Ham veri Data Lake içine taşındı.") 
```

```python
import os 

 

os.makedirs("warehouse", exist_ok=True) 

 

summary = {} 

 

with open("logs.csv", "r", encoding="utf-8") as f: 

    for line in f: 

        parts = line.strip().split(",") 

        content = parts[1] 

 

        summary[content] = summary.get(content, 0) + 1 

 

with open("warehouse/content_summary.csv", "w", encoding="utf-8") as f: 

    for k, v in summary.items(): 

        f.write(f"{k},{v}\n") 

 

print("Warehouse raporu hazırlandı.") 
```

```python
import os 

import shutil 

 

os.makedirs("lakehouse/bronze", exist_ok=True) 

os.makedirs("lakehouse/silver", exist_ok=True) 

os.makedirs("lakehouse/gold", exist_ok=True) 

 

# Bronze = ham veri 

shutil.copy("logs.csv", "lakehouse/bronze/logs.csv") 

 

# Silver = temiz veri 

clean_rows = [] 

 

with open("logs.csv", "r", encoding="utf-8") as f: 

    for line in f: 

        parts = line.strip().split(",") 

        user_id = parts[0] 

        content = parts[1] 

        watch_time = int(parts[2]) 

        device = parts[3] 

 

        if watch_time > 0: 

            clean_rows.append(line.strip()) 

 

with open("lakehouse/silver/clean_logs.csv", "w", encoding="utf-8") as f: 

    for row in clean_rows: 

        f.write(row + "\n") 

 

# Gold = özet rapor 

summary = {} 

 

for row in clean_rows: 

    parts = row.split(",") 

    content = parts[1] 

    summary[content] = summary.get(content, 0) + 1 

 

with open("lakehouse/gold/report.csv", "w", encoding="utf-8") as f: 

    for k, v in summary.items(): 

        f.write(f"{k},{v}\n") 

 

print("Lakehouse yapısı hazırlandı.") 
```
