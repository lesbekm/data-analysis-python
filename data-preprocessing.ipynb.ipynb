```python
class Ogrenci:
    def __init__(self,isim,numara,bolum):
        self.isim=isim
        self.numara=numara
        self.bolum=bolum
        self.notlar=[]
    


    def not_ekle(self,not_degeri):
        self.notlar.append(not_degeri)
   


    def ortalama_hesapla(self):
        if self.notlar:
            return sum(self.notlar)/len(self.notlar)
        return 0
    
    
    def bilgileri_goster(self):
        print(f"isim: {self.isim}")
        print(f"ortalama: {self.ortalama_hesapla()}")
   
```

```python
ogr1=Ogrenci("Ali","111","bilgisayar")
```

```python
ogr1.not_ekle(80)
ogr1.not_ekle(50)
ogr1.not_ekle(70)
ogr1.bilgileri_goster()
```
