```python
carp = lambda x, y: x * y
print(carp(4, 7))

cift_mi = lambda x: x % 2 == 0
print(cift_mi(4))
print(cift_mi(7))

buyuk = lambda s: s.upper()
print(buyuk("python"))
```

```python
func1 = lambda x: x * 2
func2 = lambda x: x + 10
func3 = lambda x: x ** 2

sonuc = func3(func2(func1(5)))

print(sonuc)
```

```python
matrix = [[1, 2], [3, 4], [5, 6]]

flat = [x * 2 for row in matrix for x in row if x % 2 == 0]

print(flat)
```

```python
metin = "Python"

sonuc = [c.lower() if c.isupper() else c.upper() for c in metin]
print(sonuc)
print(''.join(sonuc))
```

```python
liste = [x+y for x in range(3) for y in range(3) if x!=y]

print(len(liste))
print(sum(liste))
```

```python
veriler = [2, 5, 8, 12, 3]

mesaj = ["Çift" if x % 2 == 0 and x > 5 else "Küçük" if x < 4 else "Tek" for x in veriler]
print(mesaj)
```

```python
fiyatlar = [150, 25, 50, 200, 20]

yeni_fiyatlar = [f * 0.9 if f > 100 else 10 if f < 30 else f for f in fiyatlar]
print(yeni_fiyatlar)
```

```python
kullanicilar = ["ALİ", "ayŞE", "  Mehmet  ", "", "   ", "CAN"]

temiz = [k.strip() for k in kullanicilar]
print(temiz)

dolu = [k for k in temiz if k]
print(dolu)

standart = [k.capitalize() for k in dolu]
print(standart)

temiz_kullanicilar = [k.strip().capitalize() for k in kullanicilar if k.strip()]
print(temiz_kullanicilar)
```

```python
fiyatlar = [100, 250, 50, 800, 120, 1500, 75]

kdvli = list(map(lambda f: f * 1.18, fiyatlar))
print(kdvli)

indirimli = [f * 0.8 for f in fiyatlar]
print(indirimli)

pahali = list(filter(lambda x: x > 700, fiyatlar))
print(pahali)

son_fiyatlar = list(map(lambda x: x * 0.8 * 1.18, fiyatlar))
print(son_fiyatlar)
```
