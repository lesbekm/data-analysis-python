```python
import matplotlib.pyplot as plt
labels=['SciFi','Drama','Comedy','Action','Romance','Thriller']
sizes=[5,15,10,20,40,10]
explode=(0,0,0,0,0.1,0)
plt.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',
        shadow=True,startangle=98)
plt.show()
```

```python
import matplotlib.pyplot as plt

categories=["Matematik","Fizik","Kimya","Biyoloji"]
values=[85,90,78,88]

plt.bar(categories,values)

plt.title("Öğrenci puanları")
plt.xlabel("Dersler")
plt.ylabel("Puanlar")

plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np

veri=np.random.normal(70,10,100)

plt.hist(veri,bins=10)

plt.title("Notların dağılımı")
plt.xlabel("Not")
plt.ylabel("Frekans")

plt.show()
```

```python
import matplotlib.pyplot as plt

enerji_kaynakları=[40,25,20,12,10,22]

etiketler=[
    "güneş enerjisi",
    "rüzgar enerjisi",
    "hidroelektrik",
    "doğal gaz",
    "kömür",
    "ışıklar"
]

renkler=[
    "AZURE",
    "LIGHTYELLOW",
    "TAN",
    "MEDIUMPURPLE",
    "ROSYBROWN",
    "CYAN"
]

plt.pie(
enerji_kaynakları,
labels=etiketler,
autopct="%1.0f%%",
colors=renkler
)

plt.title("elektrik üretim kaynaklarının dağılımı")

plt.show()
```
